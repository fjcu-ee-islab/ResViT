```python
import matplotlib.pyplot as plt   # 
import cv2

img0=cv2.imread('heatdata/severe/PACG/00010.jpg')
img1=cv2.imread('heatdata/severe/heatmap/PACG/10gradcam_cam.jpg')
img2=cv2.imread('heatdata/severe/heatmap/PACG/10gradcam++_cam.jpg')
img3=cv2.imread('heatdata/severe/heatmap/PACG/10eigencam_cam.jpg')
img4=cv2.imread('heatdata/severe/heatmap/PACG/10eigengradcam_cam.jpg')
img5=cv2.imread('heatdata/severe/heatmap/PACG/10scorecam_cam.jpg')
img6=cv2.imread('heatdata/severe/heatmap/PACG/10layercam_cam.jpg')

b,g,r = cv2.split(img0)
img0 = cv2.merge([r,g,b])
b,g,r = cv2.split(img1)
img1 = cv2.merge([r,g,b])
b,g,r = cv2.split(img2)
img2 = cv2.merge([r,g,b])
b,g,r = cv2.split(img3)
img3 = cv2.merge([r,g,b])
b,g,r = cv2.split(img4)
img4 = cv2.merge([r,g,b])
b,g,r = cv2.split(img5)
img5 = cv2.merge([r,g,b])
b,g,r = cv2.split(img6)
img6 = cv2.merge([r,g,b])

plt.figure(figsize=(30,30))
plt.subplot(2, 4, 1)
plt.imshow(img0)
plt.title("raw",fontsize=25)
plt.subplot(2, 4, 2)
plt.imshow(img1)
plt.title("gradcam",fontsize=25)
plt.subplot(2, 4, 3)
plt.imshow(img2)
plt.title("gradcam++",fontsize=25)
plt.subplot(2, 4, 4)
plt.imshow(img3)
plt.title("eigencam",fontsize=25)
plt.subplot(2, 4, 6)
plt.imshow(img4)
plt.title("eigengradcam",fontsize=25)
plt.subplot(2, 4, 7)
plt.imshow(img5)
plt.title("scorecam",fontsize=25)
plt.subplot(2, 4, 8)
plt.imshow(img6)
plt.title("layercam",fontsize=25)
plt.subplots_adjust(left=None, bottom=None, right=None, top=None,
                wspace=None, hspace=-0.6)
plt.savefig('heatdata/severe/heatmap/all/PACG/10all.png',dpi=600, bbox_inches='tight')
```
